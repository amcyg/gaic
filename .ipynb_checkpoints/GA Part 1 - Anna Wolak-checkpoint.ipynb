### Computing mean smoothness and compactness
* The data is skewed, so the mean will be different from the median (unlike in a normal distribution, where both are equal)
* Since our sample size is rather small, we can use bootstrapping to resample with replacement
* The distribution of the means of the bootstrapped means will be normal

#### Let's look at the distributions again for smoothness and compactness in benign vs. malignant cells

```python
grouped_df = df.groupby('diagnosis')

# Smoothness
grouped_df['smoothness_mean'].plot(kind='hist', alpha=0.3, figsize=(5, 5), title='smoothness_mean', legend=True)
plt.show()

# Compactness
grouped_df['compactness_mean'].plot(kind='hist', alpha=0.3, figsize=(5,5), title='compactness_mean', legend=True)
plt.show()
```

### Steps for bootstrapping to determine the mean:
* Take a sample with replacement 
* Take the mean of that sample, and sample again for x amount of times
* Take the mean of the collected means from all samples
* Perform t-test to determine if there is a difference between malignant and benign means
    * For a 95% confidence interval, p < 0.05 to reject the null hypothesis (i.e. show a difference)

```python
def bootstrap_for_distribution(df, target_series, num_samples):
    total_means = []
    for n in range(num_samples):
        sample = df.sample(frac=1, replace=True)
        mean = sample[target_series].mean()
        total_means.append(mean)
    total_means = np.array(total_means)
    return total_means

def find_means_and_perform_ttest(total_means_1, total_means_2):
    t_statistic, p_value = scipy.stats.ttest_ind(total_means_1, total_means_2)
    print "t-statistic, p-value: ", t_statistic, p_value
    print "Benign mean: ", np.mean(total_means_benign)
    print "Malignant mean: ", np.mean(total_means_malignant)
```

### Smoothness, malignant vs. benign

```python
total_means_benign = bootstrap_for_distribution(df[df['diagnosis'] == 'B'], 'smoothness_mean', 5000)
total_means_benign.plot(kind='hist', alpha=0.3)
total_means_malignant = bootstrap_for_distribution(df[df['diagnosis'] == 'M'], 'smoothness_mean', 5000)
find_means_and_perform_ttest(total_means_benign, total_means_malignant)
```

### Compactness, malignant vs. benign

```python
total_means_benign = bootstrap_for_distribution(df[df['diagnosis'] == 'B'], 'compactness_mean', 5000)
total_means_malignant = bootstrap_for_distribution(df[df['diagnosis'] == 'M'], 'compactness_mean', 5000)
find_means_and_perform_ttest(total_means_benign, total_means_malignant)
```

### Preparing data for modeling
* All of our features are numerical, but our labels (benign vs. malignant) are categorical.
* It's best to convert them to binary 0 vs. 1 to make modeling easier.

```python
df_modeling = df.copy()
df_modeling['diagnosis'] = df['diagnosis'].map( {'B': 0, 'M': 1}).astype(int)
```

```python
df_modeling.head()
```

* Next, we separate our data into training (80%) and testing (20%) data sets.

```python
X = df_modeling.drop(['diagnosis'], axis=1).values
y = df_modeling['diagnosis'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

* We can use K-fold cross validation to determine which model performs best in multiple iterations of the training data. We'll use 10 folds in this case.

```python
# Test options and evaluation metric
num_folds = 10
seed = 7
scoring = 'accuracy'
```

```python
# Spot checking two classification algorithms, using default tuning parameters
models = []
models.append(('LR', LogisticRegression())) 
models.append(('CART', DecisionTreeClassifier())) 
```

```python
# Gathering model cross validation results
results = []
names = []
for name, model in models:
  kfold = KFold(n_splits=num_folds, random_state=seed)
  cv_results = cross_val_score(model, X_train, y_train, cv=kfold, scoring=scoring)
  results.append(cv_results)
  names.append(name)
  msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
  print(msg)
```

```python
# Visualizing performance of models 
fig = plt.figure() 
fig.suptitle('Algorithm comparison') 
ax = fig.add_subplot(111) 
plt.boxplot(results) 
ax.set_xticklabels(names) 
plt.show()
```

### Logistic Regression
#### Let's take a look at a simple logistic regression classification model to examine how our features interact. 
* __symmetry_worst and symmetry_sd_error__ 
    * These have the highest positive correlation (as the values of these two factors increase, so does the probability of diagnosis=1, or a malignant diagnosis)
* __radius_mean and concavity_worst__
    * These have the highest negative correlation (as values increase, the probability of malignant diagnosis decreases)

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, y_train) * 100, 2)
acc_log
```

```python
coeff_df = pd.DataFrame(df.columns.delete(0))
coeff_df.columns = ['Feature']
coeff_df["Correlation"] = pd.Series(logreg.coef_[0])

coeff_df.sort_values(by='Correlation', ascending=False)
```
