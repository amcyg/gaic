### In general, how do the features of benign vs. malignant cells compare?
* __Smoothness vs. Compactness__
    * Generally, benign cells in this data set have lower smoothness and compactness factors than malignant cells do. 
    * Mean factors are greater than the median factors, showing that the data distribution is skewed instead of being normalized at the center (where the mean = median).
* __Since our data is not normally distributed, we can't compare benign vs. malignant distributions with a t-test. In order to make a comparison with statistical confidence, we need to bootstrap.__
    * We can take many means of many subsets of the data set. Thanks to the central limit theorem, the means of our subsets (if we have enough of them) will eventually create a normal distribution with a particular confidence and probability.

```python
aggregations = {
    'smoothness_mean': {
        'mean' : 'mean',
        'median' : 'median'
    },
    'compactness_mean': {
        'mean' : 'mean',
        'median' : 'median'
    }
}

df.groupby('diagnosis').agg(aggregations)
```

* __Distributions of remaining features__
    * Histograms below show the distribution differences of the remaining cell image features. 
    * Some features, such as perimeter_sd_error, concavity_worst, and fractal_dimension_mean have minimal overlap between benign and malignant cells. These features are likely to be most useful for us later when building a predictive model.

```python
fig, subplots = plt.subplots(nrows=10, ncols=3, figsize=(20,45))
df_grouped = df.groupby('diagnosis')
features = ['radius_mean', 'radius_sd_error', 'radius_worst', 'texture_mean', 'texture_sd_error', 'texture_worst', 'perimeter_mean', 'perimeter_sd_error', 'perimeter_worst', 'area_mean', 'area_sd_error', 'area_worst', 'smoothness_mean', 'smoothness_sd_error', 'smoothness_worst', 'compactness_mean', 'compactness_sd_error', 'compactness_worst', 'concavity_mean', 'concavity_sd_error', 'concavity_worst', 'concave_points_mean', 'concave_points_sd_error', 'concave_points_worst', 'symmetry_mean', 'symmetry_sd_error', 'symmetry_worst', 'fractal_dimension_mean', 'fractal_dimension_sd_error', 'fractal_dimension_worst']
for index, feature in enumerate(features):
    subplot = subplots[int(index/3), index%3]
    df_grouped[feature].plot(ax=subplot, kind='hist', alpha=0.3, legend=True, title=feature)
```

### Generating bootstrap samples
* Version 1: bootstrapping benign and malignant subsections separately

```python
benign = df[df['diagnosis'] == 'B']
malignant = df[df['diagnosis'] == 'M']
```

```python
bootstrapped_benign = benign.sample(n=10000, replace=True)
bootstrapped_malignant = malignant.sample(n=10000, replace=True)
```

```python
benign['compactness_mean'].plot(kind='hist', alpha=0.5)
bootstrapped_benign['compactness_mean'].plot(kind='hist', alpha=0.2)
```

```python
malignant['compactness_mean'].plot(kind='hist', alpha=0.5)
bootstrapped_malignant['compactness_mean'].plot(kind='hist', alpha=0.2)
```

```python
bootstrapped_benign['compactness_mean'].plot(kind='hist', alpha=0.2)
bootstrapped_malignant['compactness_mean'].plot(kind='hist', alpha=0.2)
```

```python
bootstrapped_benign['compactness_mean'].mean()
```

```python
bootstrapped_benign['compactness_mean'].median()
```
