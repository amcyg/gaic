### In general, how do the features of benign vs. malignant cells compare?
* __Smoothness__
    * Benign: 2.000321 (Mean), 1.8510 (Median)
    * Malignant: 4.323929 (Mean), 3.6795 (Median)
* __Compactness__
    * Benign: 0.021438 (Mean), 0.01631 (Median)
    * Malignant: 0.032281 (Mean), 0.02859 (Median)
* Benign cells have lower smoothness and compactness factors than malignant cells do. In all cases, the mean factors are greater than the median factors, showing that the data distribution skews to the right instead of being normalized at the center. This means that the median factor values are likely to be more useful when comparing individual cells to the typical benign or malignant cell. 

```python
df.groupby('diagnosis').mean()
```

```python
df.groupby('diagnosis').median()
```

```python
aggregations = {
    'radius_mean': {
        'max' : 'max',
        'min' : 'min',
        'median' : 'median',
        'mean' : 'mean'
    }, 
    'texture_mean': {
        'max' : 'max',
        'min' : 'min',
        'median' : 'median',
        'mean' : 'mean'
    }, 
    'smoothness_mean': {
       'max' : 'max',
        'min' : 'min',
        'median' : 'median',
        'mean' : 'mean'
    }, 
     'compactness_mean': {
        'max' : 'max',
        'min' : 'min',
        'median' : 'median',
        'mean' : 'mean'
    },
}
```

```python
df.groupby('diagnosis').agg(aggregations)
```

```python
radius = sns.violinplot(x='diagnosis', y='radius_mean', data=df, hue='diagnosis')
```
